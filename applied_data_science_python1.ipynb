##Reading Data

Once you have imported ```pandas```, you can use the path of ther data to read it.

In ```pandas``` the ```head(n)``` method will return the first ```n``` lines of your data. The ```tails(n)``` method does the same for the las ```n```lines.

````
#importing pandas
import pandas as pd
import numpy as np
data_path = "/content/imports-85.data"
````

````
#reading the data in a csv format
#Note that the data does not have headers for the data, so header=None must be used
df = pd.read_csv(data_path, header=None)
df.head(5)
````

##Adding Headers

The headers for the data are available at https://archive.ics.uci.edu/ml/datasets/Automobile

Headers will help describe the data in a more readable way.

````
#all the headers in a more convenient presentation
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
````

````
df.columns = headers
df.head(5)
````

Here we utilize the ```replace()```method to eliminate the question marks and make them ```NaN``` so that the ```dropna()``` method can eliminate the missing data.

````
df1=df.replace('?',np.NaN)
````

````
df=df1.dropna(subset=["price"], axis=0)
df.head(10)
````

To show the name of each column we use ```df.columns```

````
df.columns
````

##Saving
Pandas can also save the dataset using ```df.to_scv("filepath")```. Other formats that can be read and written to are JSON, Excel, hdf, sql and more.

###Description

The ```df.describe()``` method returns an statistical description of all data in the dataframe, excluding ```NaN``` values.

````
df.describe()
````

The ```include``` argument can be used in ```describe()``` to list even the ```object``` type data:

````
df.describe(include='all')
````

We can also describe specific columns using them as indexes:

````
df[['make', 'length']].describe(include='all')
````

The ```info()```method can also be used to take some aditional information on your dataframe.

````
df.info()
````
