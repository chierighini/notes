# Data Visualization

Is a way to show complex data in a more understandable way that can be presented and explained to others in a convenient manner.

## Best Practices

According to [Darkhorse Analytics](https://www.darkhorseanalytics.com/), a company that came out of the Alberta University in Canada, less is always more:

* Less is more effective
* Less is more attractive
* Less is more impactive

In their website there are several examples of this, since the complexity of the data visualization adds nothing to the final interpretation, and therefore should be dropped completely into a more simple manner.

# `Matplotlib`

Is based off of MATLAB and composed of three layers:

* The Backend Layer
* The Artist Layer
* The Scripting Layer

[More on the subject here.](https://www.aosabook.org/en/matplotlib.html)

## Backends
There are numerous backends that can be used for `matplotlib`.

The `%matplotlib inline` backend will generate the plots as "static" and that cannot be changed unless you plot them again.

To Bypass this, in the notebook environment, such as this one, we can use `%matplotlib notebook`. This way we can make it "non-static" and change some things while the plot is an active figure.

# Visualizing with `matplotib`

```
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```
haiti = df_can.loc['Haiti', years] # passing in years 1980 - 2013 to exclude the 'total' column
haiti.head()
```

```
haiti.plot()
```

```
haiti.index = haiti.index.map(int) # let's change the index values of Haiti to type integer for plotting
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show() # need this line to show the updates made to the figure
```

```
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

# annotate the 2010 Earthquake. 
# syntax: plt.text(x, y, label)
plt.text(2000, 6000, '2010 Earthquake')

plt.show() 
```

### Correct answer

```
df_CI = df_can.loc[['India', 'China'], years]
df_CI.head()
df_CI.plot(kind='line')
```

We need to transpose the values now to get a proper plot, since the indices are placed in the x-axis and our index is `Country`

```
df_CI = df_CI.transpose()
df_CI.head()
```

```
df_CI.index = df_CI.index.map(int) # let's change the index values of df_CI to type integer for plotting
df_CI.plot(kind='line')

plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

In the haiti plot there was no need to transpose the values since it was a series instead of a dataframe. For line plots, it shouldn't have more than 5 lines, for readability.

## Exercise
Compare the trend of top 5 countries that contributed the most to immigration to Canada.

```
#first we sort
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
#then we get the top 5 entries
df_top = df_can.head(5)
#then we transpose
df_top = df_top[years].transpose() 

print(df_top)
```

Now we work on the index:

```
df_top.index = df_top.index.map(int)
```

```
df_top.plot(kind='line',figsize=(14,8))
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```
